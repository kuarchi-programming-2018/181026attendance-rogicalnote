# ライブラリの`import`を用いる
## matplotlibを使ってグラフをプロット

まずはデータを読み込み，それをX, Yのデータにして返す関数を作りましょう。
csv とはcomma separated valuesです。テキストで開いてみればわかります。エクセルでも開けます。

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name, 'r'))
    X = []
    Y = []
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X, Y  # データを返す
```

matplotlibを用いてデータを散布図にプロットする関数を作成します。

```python
import matplotlib.pyplot as plt
% matplotlib inline
#  % matplotlib inlineはjupyter notebookでmatplotlibを行内に表示するためのおまじない
```

```python
def draw_graph(X, Y, title, xlabel, ylabel):
    import numpy as np# ここに描画するプログラムを書く。
    fig = plt.figure()
    ax = plt.subplot(1,1,1)
    X = np.linspace(-2,6,10)
    ax.set_title('2-2-4')# グラフタイトル
    ax.set_xlabel('x')# X軸タイトル
    ax.set_ylabel('y')# Y軸タイトル
    ax.grid(True)
    ax.plot(X, Y, color='black', linestyle='solid') # グラフをメモリ上に作成
    plt.show()  # グラフの描画
```

では，データを読み込んで描画して見ましょう。

```python
file_name = 'data3.csv'  # ファイル名
x,y =read_data(file_name)# 関数read_dataによりデータを取得
title = ('graph224')# グラフタイトル
xlabel = ('Xl')# X軸タイトル
ylabel = ('Yl')# Y軸タイトル
# タイトル，ラベル
draw_graph(x, y, title, xlabel, ylabel)# 関数draw_graphによりグラフ描画
```
